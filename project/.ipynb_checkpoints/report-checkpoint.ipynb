#### Data Engineering

My whole attention was on maintaining the relevance and quality of the data during the data engineering process. Using this method, I was able to create a solid dataset that was both comprehensive and ready to show the important connections between UnEmployment East and UnEmployment West Germany in Germany.

```python
# Retrieve the table names and respective data
cursor = conn.cursor()
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
table_names = cursor.fetchall()


dataframes = {}
for table in table_names:
    table_name = table[0]
    query = f"SELECT * FROM {table_name};"
    df = pd.read_sql_query(query, conn)
    dataframes[table_name] = df

#unemployed datasets
unemployed_east = dataframes['unemployed_east']
unemployed_west = dataframes['unemployed_west']
```

#### Monthly Unemployment Comparison between East and West Germany

```python
unemployed_east['date'] = pd.to_datetime(unemployed_east['year'].astype(str) + '-' + unemployed_east['month'].astype(str))
unemployed_west['date'] = pd.to_datetime(unemployed_west['year'].astype(str) + '-' + unemployed_west['month'].astype(str))

unemployed_east.set_index('date', inplace=True)
unemployed_west.set_index('date', inplace=True)

east_unemployment = (unemployed_east['male'] + unemployed_east['female']).resample('M').mean()
west_unemployment = (unemployed_west['male'] + unemployed_west['female']).resample('M').mean()

month_year_labels = [date.strftime('%Y-%b') for date in east_unemployment.index]


fig, ax1 = plt.subplots(figsize=(15, 8))

ax1.bar(range(len(east_unemployment)), east_unemployment.values, color='b', alpha=0.6, label='Average Unemployment in East Germany')
ax2 = ax1.twinx()
ax2.plot(range(len(west_unemployment)), west_unemployment.values, color='r', label='Average Unemployment in West Germany', marker='o')

ax1.set_xlabel('Month - Year',fontsize=12)
ax1.set_ylabel('Average Unemployment in East Germany', color='b')
ax2.set_ylabel('Average Unemployment in West Germany', color='r')
plt.title('Monthly Unemployment Comparison between East and West Germany')

ax1.set_xticks(range(len(month_year_labels)))
ax1.set_xticklabels(month_year_labels, rotation=45)

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

```python
# Converting 'year' and 'month' to datetime and create a new column 'month_year'
unemployed_east['month_year'] = pd.to_datetime(unemployed_east['year'].astype(str) + '-' + unemployed_east['month'].astype(str))
unemployed_west['month_year'] = pd.to_datetime(unemployed_west['year'].astype(str) + '-' + unemployed_west['month'].astype(str))

combined_df =  pd.concat([unemployed_east, unemployed_west])

combined_df['male'] = pd.to_numeric(combined_df['male'])
combined_df['female'] = pd.to_numeric(combined_df['female'])

grouped = combined_df.groupby(['month_year']).agg({'male': 'mean', 'female': 'mean', 
                                                   'male': 'mean', 'female': 'mean'}).reset_index()
fig, ax = plt.subplots(figsize=(15, 8))

bar_width = 0.2

r1 = np.arange(len(grouped))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]
r4 = [x + bar_width for x in r3]

ax.bar(r1, grouped['male'], color='b', width=bar_width, edgecolor='grey', label='Male East')
ax.bar(r2, grouped['female'], color='c', width=bar_width, edgecolor='grey', label='Female East')
ax.bar(r3, grouped['male'], color='r', width=bar_width, edgecolor='grey', label='Male West')
ax.bar(r4, grouped['female'], color='m', width=bar_width, edgecolor='grey', label='Female West')

ax.set_xlabel('Month-Year', fontsize=12)
ax.set_ylabel('Average Unemployment', fontsize=12)
ax.set_title('Comparison of Monthly Unemployment Rates (2018-2023)', fontsize=14)
ax.set_xticks([r + bar_width for r in range(len(grouped))])
ax.set_xticklabels([date.strftime('%Y-%b') for date in grouped['month_year']], rotation=45)

ax.legend()

plt.tight_layout()
plt.show()
```

#### Predictive analysis of unemployment data in East and West Germany:

A time series forecasting strategy is one that we may think about. Since our data is a time series consisting of monthly data points collected over a number of years, this makes sense. Forecasting time series with ARIMA (AutoRegressive Integrated Moving Average) models is a popular and efficient technique. Utilizing past data, the objective here may be to predict future rates of unemployment. We will anticipate the overall unemployment rates (male + female) for both East and West Germany separately.

```python
# preprocessing data for both regions
def preprocess_data(df):
    df['date'] = pd.to_datetime(df[['year', 'month']].assign(DAY=1))
    df.set_index('date', inplace=True)
    df['total_unemployment'] = df['male'] + df['female']
    df.sort_values(by=['year', 'month'], inplace=True)
    return df.asfreq('MS')

def fit_predict_model(data):
    train_size = int(len(data) * 0.8)
    train, test = data[0:train_size], data[train_size:len(data)]
    model = ExponentialSmoothing(train, seasonal='add', seasonal_periods=12)
    model_fit = model.fit()
    predictions = model_fit.forecast(steps=len(test))
    return train, test, predictions


# Applying preprocessing
df_east = preprocess_data(unemployed_east)
df_west = preprocess_data(unemployed_west)


combined_data = pd.concat([df_east, df_west])

# Group by date and sum the total unemployment
combined_data = combined_data.groupby('date')['total_unemployment'].sum().asfreq('MS')

# Fit and predict using the combined data
train_combined, test_combined, predictions_combined = fit_predict_model(combined_data)

# Plotting for combined data
plt.figure(figsize=(12,6))
plt.plot(train_combined.index, train_combined, label='Train - Combined')
plt.plot(test_combined.index, test_combined, label='Test - Combined')
plt.plot(test_combined.index, predictions_combined, label='Predictions - Combined', color='purple')
plt.title('Combined Unemployment Prediction for East and West Germany')
plt.legend()
plt.show()
```
